```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("../archive/movies_recommendation_data/tmdb_5000_movies.csv")
credits = pd.read_csv("../archive/movies_recommendation_data/tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
movies = movies.merge(credits, on="title")
```

```python
movies.columns
```

```python
"""
genre
id 
keywords
overview
original_language
runtime
cast
"""

movies = movies[["movie_id", "title","overview", "genres", "keywords", "cast", "crew" ]]
```

```python
movies.isnull().sum()
```

```python
movies.dropna()
```

```python
join_words = lambda string: string.replace(" ", "")
```

```python
def convert_genres(row):
    row = eval(row)
    row = [genre["name"].replace(" ", "") for genre in row]
    return " ".join(row)

movies["genres"] = movies["genres"].apply(convert_genres)
```

```python
movies["keywords"] = movies["keywords"].apply(convert_genres)
```

```python
movies
```

```python
movies["cast"][0]
```

```python
type(movies['cast'][0])
```

```python
def list_actor_charaters(row, key, offset=3):
    row = eval(row)
    row = row[:offset]
    row = [cast[key].replace(" " , "") for cast in row]
    return " ".join(row)

movies['character'] = movies['cast'].apply(list_actor_charaters, args=("character",))
movies['cast'] = movies['cast'].apply(list_actor_charaters, args=("name",))
```

```python
def get_director(row):
    row = eval(row)
    return next((crew["name"] for crew in row if crew['job'] == "Director"), None)

movies["crew"] = movies["crew"].apply(get_director)
```

```python
movies.head()
```

```python
cols = ["overview", "genres", "crew", "cast", "character"]

movies["tags"] = (
    movies[cols]
    .fillna("")
    .agg(
        lambda x: 
        f" overview {x['overview']}"
        f" \ngenres {x['genres']}"
        f" \ncrew {x['crew']}"
        f" \ncast {x['cast']}"
        f" \ncharacter {x['character']}",
        axis=1
    )
)
```

```python
movies["tags"] = movies["tags"].str.lower()
```

```python
movies.head()
```

```python
movies[movies.isnull().any(axis=1)]
```

```python
movies_new = movies[["movie_id","title","tags"]]
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
docs = nlp.pipe(movies_new["tags"], batch_size=1000)
movies_new["tags"] = [
    " ".join(t.lemma_ for t in doc if t.is_alpha)
    for doc in docs
]
```

```python
movies_new[movies_new.isnull().any(axis=1)]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(movies_new["tags"]).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie, n=5):
    if movie not in movies_new['title'].values:
        print("Movie not found")
        return

    index = movies_new.index[movies_new['title'] == movie][0]
    distances = similarity[index]
    top_indices = np.argsort(distances)[-n-1:-1][::-1]
    for i in top_indices:
        print(f"title: {movies_new.iloc[i].title}")
```

```python
recommend("Batman Begins")
```
